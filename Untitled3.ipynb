```python

 import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

y = (ipl_clean['is_captain'] == 'Yes').astype(int)

X = ipl_clean[['matches', 'runs', 'avg', 'strike_rate', 'fifties', 'hundreds']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

models = {
    "Mahendra Singh Dhoni": LogisticRegression(max_iter=10),
    "Ruturaj Gaikwad": KNeighborsClassifier(n_neighbors=20),
    "Sanju Samson": DecisionTreeClassifier(max_depth=25, random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc

    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot()
    plt.title(f"{name} Confusion Matrix")
    plt.show()

best_model = max(results, key=results.get)
print(f"THE CAPTAIN FOR CSK TEAM IN 2027 IS {best_model} WITH ACCURACY {results[best_model]:.2f} .")
```
