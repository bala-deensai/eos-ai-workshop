```python
import numpy as np
import matplotlib.pyplot as plt

p = np.zeros((16,16))

# car body
p[8:11, 3:13] = 1

# roof
p[6:8, 6:10] = 1

# bigger wheels (2x2 blocks)
p[11:13, 4:6] = 1   # left wheel
p[11:13, 10:12] = 1 # right wheel

# display
plt.imshow(p, cmap='gray')
plt.axis('image')
plt.axis('off')
plt.show()
```
